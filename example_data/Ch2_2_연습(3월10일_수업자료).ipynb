### Example 2.2.2   numpy

```python
import numpy as np
import math

# 근의공식 구하기
# a가 0이 아니고 2개의 해가 있다고 가정
a = int(input('a값을 입력하고 Enter: '))
b = int(input('b값을 입력하고 Enter: '))
c = int(input('c값을 입력하고 Enter: '))

x1 = ( (-b + (math.sqrt(b**2 - 4 * a * c)))/2 * a)
x2 = ( (-b - (math.sqrt(b**2 - 4 * a * c)))/2 * a)

print('x1 = ' , x1)
print('x2 = ' , x2)
```

### Example 2.2.4       횟수 반복문

```python
# 0부터 4까지 5번 반복

for i in range(5):
   print('iteration')
```

```python
# 0부터 4까지 5번 반복하며 i값 출력

for i in range(5):
   print('iteration', i)
```

```python
# 1부터 9까지 5번 반복 (홀수번)

for i in range(1, 10, 2):
   print('iteration', i)
```

### Example 2.2.5   while 반복문

```python
# for 문

sum = 0
for i in range(1, 11):
   sum = sum + i
print('합계=', sum)


#while 문
i = 1
sum = 0
while i < 11:
   sum = sum + i
   i = i + 1
print('합계=', sum)
```

```python
 # 주어진 숫자만큼 문자를 반복
def rept(string, n):
   n_st=''
   for i in range(n):
     n_st = n_st + string
   print(n_st)

string = input('문자를 입력하고 Enter: ')
n = int(input('정수숫자를 입력하고 Enter: '))

rept(string, n)
```

### Example 2.2.7

```python
# 주어진 두 숫자의 평균을 반환하는 함수
def avr(a, b):
  sum = a + b
  return sum/2

a = int(input('정수 a를 입력하고 Enter: '))
b = int(input('정수 b를 입력하고 Enter: '))

result = avr(a, b)
print(result)
```
