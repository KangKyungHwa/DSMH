## 4. feedparser를 이용하여 뉴스 데이터 수집하기
#### (1)RSS/Atom 같은 웹 피드(XML 기반)를 파싱해서 파이썬 객체로 바꿔주는 라이브러리
#### (2)파이썬 딕셔너리 형태로 변환해줌

```python
import feedparser
import requests

# Google News RSS 피드 URL을 사용합니다.
url = "https://news.google.com/rss?hl=ko&gl=KR&ceid=KR:ko"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/121.0.0.0 Safari/537.36",
    "Accept": "application/rss+xml, application/xml;q=0.9,*/*;q=0.8"
}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status() # HTTP 오류가 발생하면 예외를 발생시킵니다.
    feed = feedparser.parse(response.text) # 가져온 텍스트 내용을 파싱합니다.

    if not feed.feed:
        print("오류: RSS 피드 메타데이터를 찾을 수 없습니다.")
    else:
        print("피드 제목:", feed.feed.get('title', '제목 없음'))

    if not feed.entries:
        print("오류: 피드에서 기사를 찾을 수 없습니다.")
    else:
        print(f"파싱된 뉴스 기사의 개수: {len(feed.entries)}")
        print("\n--- 주요 뉴스 기사 제목 (상위 5개) ---")
        for i, entry in enumerate(feed.entries[:100]):
            print(f"{i+1}. {entry.get('title', '제목 없음')}")

except requests.exceptions.RequestException as e:
    print(f"요청 중 오류 발생: {e}")
except Exception as e:
    print(f"피드 파싱 중 오류 발생: {e}")
```

###5. 불러온 기사들을  데이터프레임에 저장하기

```python
import pandas as pd

# DataFrame에 저장할 데이터를 담을 리스트를 생성합니다.
news_data = []

# feed.entries를 반복하여 각 기사의 정보를 추출합니다.
for entry in feed.entries:
    title = entry.get('title', '제목 없음')
    link = entry.get('link', '링크 없음')
    published = entry.get('published', '발행일 정보 없음')
    summary = entry.get('summary', '요약 정보 없음')

    news_data.append({
        '제목': title,
        '링크': link,
        '발행일': published,
        '요약': summary
    })

# 추출된 데이터를 사용하여 pandas DataFrame을 생성합니다.
df_news = pd.DataFrame(news_data)

# 생성된 DataFrame의 첫 5행을 출력하여 확인합니다.
print("--- 뉴스 기사 DataFrame (상위 5개) ---")
display(df_news.head())

# DataFrame의 총 행 수를 출력합니다.
print(f"\nDataFrame에 저장된 총 뉴스 기사 수: {len(df_news)}개")
```

###6. 수집된 기사 파일로 저장하기

```python
#한글이 저장될 때 깨지지 않도록 utf-8 로 인코딩해서 저장
df_news.to_csv('google_news_articles.csv', index=False, encoding='utf-8')

print("DataFrame이 'google_news_articles.csv' 파일로 성공적으로 저장되었습니다.")
print("왼쪽 파일 탐색기에서 파일을 확인할 수 있습니다.")
```
