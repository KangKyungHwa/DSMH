## Ch 4.2.







```python
Lst1 = [1, 3, 5, 7]
Lst2 = [2, 4, 6, 8, 10, 12]

print(Lst1, '\n')
print(Lst2, '\n')

print(Lst1.index(7), '\n')
print(Lst1, '\n')

Lst1.append(9)   #Lst1에 9 추가
print(Lst1, '\n')

Lst1.insert(0, 11) #Index 0의 자리에 11 추가
print(Lst1, '\n')
```

```python
Lst1 = [1, 3, 5, 7]
Lst2 = [2, 4, 6, 8, 10, 12]

Lst2.remove(12)    #Lst2 의 12 삭제
print(Lst2, '\n')

Lst1.extend(Lst2)   #Lst1 뒤에 Lst2  추가하기
print(Lst1, '\n')

Lst1.sort()         # 정렬
print(Lst1, '\n')

Lst1.reverse()      # 역으로 정렬
print(Lst1, '\n')

print(Lst1+Lst2)
```

## Ch 4.3   Numpy 라이브러리 사용



```python
import numpy as np
arr1=np.array([1, 2, 3, 4, 5])
arr2=np.array([6, 7, 8, 9, 10])
print(arr1)

print(arr1+arr2)    #파이선의 LIST와 다름 : print(Lst1+Lst2)

print(arr1+5)

print(arr1-arr2)

print(arr1-5)

print(arr1*arr2)

print(arr1*2)

print(arr1/arr2)

print(arr1/5)
```

## Ch 4.5.   Numpy 의 배열 4칙 연산


```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

# 배열 덧셈 c = a + b
c = np.add(a, b)
print(c)

# 배열 뺄셈 c = a - b
c = np.subtract(a, b)
print(c)

# 배열 곱셈 c = a * b
c = np.multiply(a, b)
print(c)

# 배열 나눗셈 c = a / b
c = np.divide(a, b)
print(c)
```

```python
# 프로그램 3
import pandas as pd
df1 = pd.DataFrame(
        [[1, 4],
         [2, 5],
         [3, 6]],
        columns=['a', 'b'])
print(df1, '\n')

df2 = pd.DataFrame(
        {"c" : [7, 8, 9],
         "d" : [10, 11, 12]})
print(df2, '\n')

df4 = pd.concat([df1, df1], axis=0)
print(df4, '\n')

df4 = pd.concat([df1, df1], axis=1)
print(df4, '\n')

df5= pd.concat([df1, df2], axis=0)
print(df5, '\n')

df6 = pd.concat([df1, df2], axis=1)
print(df6, '\n')
```

## Ch 4.12.



```python
import pandas as pd
df_x = pd.DataFrame(
        [['A', 1],
         ['B', 2],
         ['C', 3]],
        columns=['x1', 'x2'])
print(df_x, '\n')

df_y = pd.DataFrame(
        {"x1" : ['B', 'C', 'D'],
         "x3" : [2, 3, 4]})
print(df_y, '\n')

print(pd.merge(df_x, df_y, how='left'), '\n')

print(pd.merge(df_x, df_y, how='right'), '\n')

print(pd.merge(df_y, df_x))
```

## Ch 4.13.



```python
import pandas as pd
df = pd.DataFrame(
   [['20225001','A', 77],
   ['20225001','B', 80],
   ['20225002','A', 85],
   ['20225002','B', 82],
   ['20225003','A', 95],
   ['20225003','B', 90]],
    columns=['ID_num','Subject', 'Score'])
print(df, '\n')

df_pivot = df.pivot(index = 'ID_num',
        columns='Subject', values = 'Score')
print(df_pivot, '\n')

df_pivot = df.pivot(index='ID_num', columns='Subject').sum(axis=1)
print(df_pivot, '\n')
```

## Ch 4.24.  fillna()   NaN (Not a Number) 값들을 특정 값으로 채워줌
