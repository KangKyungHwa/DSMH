### (1) 결과값을 보기 쉽게 만들기

```python
import requests
import datetime
import json

url="http://api.open-notify.org/iss-now.json"
r=requests.get(url)

data = r.json()

lat = data['iss_position']['latitude']
lon = data['iss_position']['longitude']

# Corrected: Access 'timestamp' directly from the 'data' dictionary
timestamp_value = data['timestamp']
dt = datetime.datetime.fromtimestamp(timestamp_value, datetime.timezone.utc)

print("위도:", lat)
print("경도:", lon)
print("시간 (UTC):", dt)  # UTC 기준 날짜/시간
```

### (2) 위도와 경도값을 지도로 표시하기

```python
import requests
import datetime
import folium

# 1️⃣ ISS 현재 위치 가져오기
url = "http://api.open-notify.org/iss-now.json"
r = requests.get(url)
data = r.json()

lat = float(data['iss_position']['latitude'])
lon = float(data['iss_position']['longitude'])

# 2️⃣ timestamp → 한국 시간 변환
timestamp = data['timestamp']
dt_utc = datetime.datetime.fromtimestamp(timestamp, datetime.timezone.utc)
dt_kst = dt_utc + datetime.timedelta(hours=9)

# 3️⃣ folium 지도 생성     #zoom_start:지도의 처음 확대.ㅊ
m = folium.Map(location=[lat, lon], zoom_start=3)

# 4️⃣ ISS 위치 마커 추가
popup_text = f"ISS 위치\n위도: {lat}\n경도: {lon}\n한국 시간: {dt_kst.strftime('%Y-%m-%d %H:%M:%S')}"
folium.Marker([lat, lon], popup=popup_text).add_to(m)

# 5️⃣ 지도 저장
m.save("iss_map.html")
print("지도 생성 완료! → iss_map.html 파일 확인")  #현재 사용하는 folder안에 생성
```

### (4) excel 파일로 저장하기

```python
import pandas as pd
import requests
import json
import time

url="http://api.open-notify.org/iss-now.json"

# 데이터를 저장할 리스트 생성
iss_positions = []

for i in range(5):
    r=requests.get(url)
    dt_dict =r.json()
    position = dt_dict['iss_position']
    print(position)
    iss_positions.append(position) # 위치 데이터를 리스트에 추가
    time.sleep(5)   # 5초에 한 번 씩 10번  나타내기

# 리스트로부터 DataFrame 생성
df = pd.DataFrame(iss_positions)

# DataFrame을 Excel 파일로 저장
# 'excel_writer' 인수는 파일 경로를 지정합니다.
# index=False로 설정하여 인덱스를 파일에 저장하지 않음
df.to_excel('iss_position.xlsx', index=False)
print("ISS 위치 데이터가 'iss_position.xlsx' 파일로 저장되었습니다.")
```
