##5번 반복 프로그램

```python
import pandas as pd
import requests
import json
import time

url="http://api.open-notify.org/iss-now.json"

# 데이터를 저장할 리스트 생성
iss_positions = []


while True:          #무한반복이므로 Ctrl +C로 강제종료
#for i in range(5):
    r=requests.get(url)
    dt_dict =r.json()
    position = dt_dict['iss_position']
    print(position)
    iss_positions.append(position) # 위치 데이터를 리스트에 추가
    time.sleep(5)   # 5초에 한 번 씩 10번  나타내기

# 리스트로부터 DataFrame 생성
df = pd.DataFrame(iss_positions)

# DataFrame을 Excel 파일로 저장
# 'excel_writer' 인수는 파일 경로를 지정합니다.
# index=False로 설정하여 인덱스를 파일에 저장하지 않음
df.to_excel('iss_position.xlsx', index=False)
print("ISS 위치 데이터가 'iss_position.xlsx' 파일로 저장되었습니다.")
```

##무한 반복되는되는 위도와 경도를 지도위에 나타내기

```python
import folium

# iss_positions 리스트는 이미 커널에 존재합니다.
# iss_positions = [{'longitude': '-29.7416', 'latitude': '-29.3791'}, {'longitude': '-29.4977', 'latitude': '-29.6057'}, {'longitude': '-29.2280', 'latitude': '-29.8545'}, {'longitude': '-28.9816', 'latitude': '-30.0800'}, {'longitude': '-28.6844', 'latitude': '-30.3499'}, {'longitude': '-28.4353', 'latitude': '-30.5742'}, {'longitude': '-28.1849', 'latitude': '-30.7980'}]

if iss_positions:
    # 첫 번째 위치를 지도의 중심으로 설정
    first_lat = float(iss_positions[0]['latitude'])
    first_lon = float(iss_positions[0]['longitude'])
    m = folium.Map(location=[first_lat, first_lon], zoom_start=2)

    # 모든 ISS 위치에 마커 추가
    for i, position in enumerate(iss_positions):
        lat = float(position['latitude'])
        lon = float(position['longitude'])
        folium.Marker(
            location=[lat, lon],
            popup=f"ISS Position {i+1}<br>Lat: {lat:.4f}, Lon: {lon:.4f}",
            icon=folium.Icon(color='red')
        ).add_to(m)

    # 지도를 HTML 파일로 저장
    map_filename = 'iss_position_map.html'
    m.save(map_filename)
    print(f"ISS 위치 지도가 '{map_filename}' 파일로 저장되었습니다.")
    # Colab에서 파일을 다운로드할 수 있도록 링크를 제공합니다.
    from google.colab import files
    files.download(map_filename)
else:
    print("수집된 ISS 위치 데이터가 없습니다. 먼저 데이터 수집 코드를 실행해주세요.")
```
